# Pega a cotação das ações em Python

```python
# Importando as bibliotecas
from pandas_datareader import data as web # Biblioteca que vai ler o arquivo da internet
import pandas as pd
import matplotlib.pyplot as plt # Biblioteca para montagem de gráficos
```

```python
# Cria um df com o valor da cotação da bovespa
cotacao_bovespa = web.DataReader('^BVSP', data_source='yahoo', start='01/01/2021', end='12/31/2021')
display(cotacao_bovespa)
```

```python
# Montar o gráfico com a coluna Adj Close
cotacao_bovespa['Adj Close'].plot(figsize=(15,10))
# Mostra o gráfico
plt.show()
```

```python
# Cria um df com o valor da cotação da bovespa
cotacao_petr = web.DataReader('PETR4.SA', data_source='yahoo', start='01/01/2021', end='12/31/2021')
# Montar o gráfico com a coluna Adj Close
cotacao_petr['Adj Close'].plot(figsize=(15,10))
# Mostra o gráfico
plt.show()
```

```python
# Tabela com a lista de empresas que você quer obter as cotações
tabela_empresa = pd.read_excel('Empresas.xlsx')
display(tabela_empresa)
```

```python
# Lê a tabela e para cada empresa, monta o gráfico
for empresa in tabela_empresa['Empresas']:
    # Mostra o nome da empre
    print(empresa)
    cotacao = web.DataReader(f'{empresa}.SA', data_source='yahoo', start='01/01/2021', end='12/31/2021')
    # Montar o gráfico com a coluna Adj Close
    cotacao['Adj Close'].plot(figsize=(20,15),label=empresa)
# Mostra o gráfico
plt.legend()
plt.show()
```
