```python
# Importando as bibliotecas
import pandas as pd

# Criar dataframe
carteira = pd.read_excel('Carteira.xlsx')
display(carteira)
```

```python
from alpha_vantage.timeseries import TimeSeries
chave_api = 'VJOW28209WPEI9F1'

ts = TimeSeries(key=chave_api, output_format='pandas')
data_inicial = '2021-01-05'

tabela_acoes = {}
display(ts)
```

```python
for acao in carteira['Ativos']:
    data, meta_data = ts.get_daily_adjusted(f'{acao}.SAO', outputsize='full')
    tabela_acoes[acao]=data
    carteira.loc[carteira['Ativos']==acao, 'Qtde'] = carteira.loc[carteira['Ativos']==acao, 'Início'].values / data.loc[data_inicial, '5. adjusted close'].values
    
    display(carteira)
    display(tabela_acoes['ITUB4'])
```

```python
data, meta_data = ts.get_daily_adjusted('BOVA11.SAO', outputsize='full')
ibov = data
display(ibov)
```

```python
data_final = '2021-12-08'
ibov = ibov.loc[data_inicial:data_final, '5. adjusted close']
ibov.plot()
```

```python
tabela_cotacao = pd.DataFrame()
for acao in tabela_acoes:
    tabela_cotacao[acao] = tabela_acoes[acao].loc[data_inicial:data_final, '5. adjusted close'] 
display(tabela_cotacao)
```

```python
for acao in tabela_cotacao.columns:
    tabela_cotacao[acao] = tabela_cotacao[acao]*carteira.loc[carteira['Ativos'] == acao,'Qtde'].values
display(tabela_cotacao)
```

```python
tabela_cotacao['Total'] = tabela_cotacao.sum(axis=1)
display(tabela_cotacao)
```

```python
carteira_ajustada = tabela_cotacao['Total']/tabela_cotacao['Total'].iloc[-1]
ibov_ajustado = ibov/ibov.iloc[-1]
carteira_ITUB4 = tabela_cotacao['ITUB4']/tabela_cotacao['ITUB4'].iloc[-1]
carteira_ABEV3 = tabela_cotacao['ABEV3']/tabela_cotacao['ABEV3'].iloc[-1]
carteira_VALE3 = tabela_cotacao['VALE3']/tabela_cotacao['VALE3'].iloc[-1]
carteira_BBDC3 = tabela_cotacao['BBDC3']/tabela_cotacao['BBDC3'].iloc[-1]

import matplotlib.pyplot as plt


carteira_ajustada.plot(label='Carteira',figsize=(15,10))
ibov_ajustado.plot(label='IBOV',figsize=(15,10))
carteira_ITUB4.plot(label='ITUB4',figsize=(15,10))
carteira_ABEV3.plot(label='ABEV3',figsize=(15,10))
carteira_VALE3.plot(label='VALE3',figsize=(15,10))
carteira_BBDC3.plot(label='BBDC3',figsize=(15,10))

plt.legend()
```
