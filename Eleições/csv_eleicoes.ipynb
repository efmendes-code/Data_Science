```python
# Vamos carregar o arquivo com o resultado da eleição no brasil dividido por municipio/zona
df_votacao = pd.read_csv('votacao_candidato_munzona_2018_BR.csv', sep=';', encoding='latin1')
df_votacao.head()
```

```python
# Visualizando o dataframe transposto
df_votacao.head().T
```

```python
# Separar os dados pois precisamos apenas dos dados do 2º turno
df_votacao = df_votacao[df_votacao['NR_TURNO']==2]
df_votacao.sample(3).T
```

```python
# Vamos verificar quantos municipios unicos existem no conjunto de dados
# Vamos pegar por codigo do municipio
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Mas o Brasil tem 5570 municipios
# Vamos verificar algumas cidades fora do Brasil
df_votacao[df_votacao['SG_UF']=='ZZ'].sample(4).T
```

```python
# Quantas cidade fora do Brasil estão no conjunto de dados?
df_votacao[df_votacao['SG_UF']=='ZZ']['CD_MUNICIPIO'].nunique()
```

```python
# Vamos eliminar os municipios de fora do Brasil
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```python
# Verificando a quantidade de municipios
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separados por municipio/zona
# Municipios pequenos normalmente tem apenas 1 zona eleitoral
# mas municipios vamaores podem ter várias zonas eleitorais

# Vamos verificar Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO']=='BELO HORIZONTE']
```

```python
# Eliminar algumas colunas do dataframe que não serão utilizadas
df_votacao = df_votacao[{'SG_UF','CD_MUNICIPIO','NM_MUNICIPIO','NR_CANDIDATO',
                       'NM_URNA_CANDIDATO','QT_VOTOS_NOMINAIS'}]
df_votacao.head()
```

```python
# Precisamos agragar os dados por municipio e candidato
df_votacao.groupby(['SG_UF','CD_MUNICIPIO','NM_MUNICIPIO','NR_CANDIDATO',
                       'NM_URNA_CANDIDATO']).sum()
```

```python
# Fazer o groupby, reindexar o indixe, vai salbvar isso em um dataframa
df_votacao = df_votacao.groupby(['SG_UF','CD_MUNICIPIO','NM_MUNICIPIO','NR_CANDIDATO',
                       'NM_URNA_CANDIDATO']).sum().reset_index()
df_votacao.sample(4)
```

```python
# Vamos verificar novamente a cidade de Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO']=='BELO HORIZONTE']
```

```python
# Ordenar o dataframe pela quantidade de votos e depois deixar apenas o primeiro 
# registro de cada municipio
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS',ascending=False).head(10)
```

```python
# Vamos ordenar
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS',ascending=False, inplace=True)

# Dropar as linhas duplicadas, matendo a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)
df_votacao.head(10)
```

```python
# Vamos verificar novamente a cidade de Belo Horizonte
df_votacao[df_votacao['NM_MUNICIPIO']=='BELO HORIZONTE']
```

```python
# Agora vamos criar uma coluna representando o resultado da eleição naquele municipio
# sendo que o resultado é true se o candidato bolsonaro venceu naquele municipio
# o contrario é falso
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# Verificando o resultado
df_votacao.head(10)
```

```python
# Mostrando alguns valores do dataframe do geopandas
df_mapa.sample(3)
```

```python
# Tentando encontrar o municipio pelo codigo (codigo dos municipios são diferentes)
df_votacao[df_votacao['CD_MUNICIPIO']==3156700]
```

```python
# Por que não podemos usar o nome do municipio para juntar
df_votacao[df_votacao['NM_MUNICIPIO']=='OURO BRANCO']
```

```python
# Recuperar a tabela de correspondência entre o IBGE e o TSE 
# direto do github do estadao
link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia = pd.read_csv(link)

df_equivalencia.head()
```

```python
# Vamos preparar os dataframes 
```

```python
# Vamos reindexar o dataframe de votação pelo codigo do TSE
df_vot_novo = df_votacao.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_vot_novo.head()
```

```python
# Vamos também reindexar o datafram de equivalencia pelo codigo do TSE
df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```python
# Verificando o tamanho dos dataframes
df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
# Criando um novo dataframe com base no merge e votacao de equivalencia
df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```python
# Vamos juntar o dataframe do geopandas com o dataframe votação/equivalência
df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# Transformar a coluna de codigo de IBGE do dataframe de cotacao/equivalencia
df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
# Vamos mudar o indice do dataframe mapa
df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)

df_mapa.head()
```

```python
# Vamos mudar o indice do dataframe votacao/equivalencia
df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)

df_vot_equi.head()
```

```python
# Vamos unificar o dataframe do mapa com o dataframe de votação/equivalencia
df_mapa_novo = df_mapa.join(df_vot_equi)

df_mapa_novo.head().T
```

```python
# Verificando o tamanho do mapa novo
df_mapa_novo.info()
```

```python
# Vamos descobrir os dois municipios a mais no mapa
df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
# Matar a coluna de ajuste
df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
# Vamos eliminar a informação das duas lagoas do RS
df_mapa_novo.dropna(inplace=True)
df_mapa_novo.info()
```

```python
df_mapa_novo
```

```python
# Vamos desenhar o mapa

# Criar a referência para o código de cor (false/0=red, True/1=green)
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red','green'])

# Desenhando o mapa
fig, ax = plt.subplots(1, figsize=(12, 12))

ax = df_mapa_novo.plot(column='RESULTADO',cmap=cmap, legend=True, linewidth=0.1, ax=ax,
                      edgecolor='grey')

# Escondendo os eixos
ax.set_axis_off()

# Colocando um título no mapa
ax.set_title('Resultado Eleição 2018 por Município / 2º Turno', fontdict={'fontsize':25, 'fontweight':3})

# Criando uma nota de rodapé
ax.annotate('Fonte: IBGE,2015; TSE,2019; Jornal o Estado de São Paulo, 2019', xy=(0.1, 0.08), xycoords='figure fraction',
            horizontalalignment='left', verticalalignment='top', fontsize=10, color='#555555')

# Altarando a legenda
leg = ax.get_legend()
leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
