# GRÁFICOS INTERATIVOS (DINÂMICOS) no Python com o Matplotlib

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gera os dados dos eixos x e y
x = np.arange(1, 6)
```

```python
# Altarando o estido do gráfio
plt.style.use('ggplot')

# Ligando o método interativo
plt.ion()

for i in range(10):
    # Atualiza os dados e plota o segundo gráfico
    dados = np.random.randint(0, 30, 5) # Gera um numero aleatório entre 20 e 30 - 5 vezes
    
    # limpar os dados e o gráfico
    plt.cla()
    plt.clf()
 
    # altera o tamanho do gráfico
    plt.figure(figsize=(20,15))
    
    # Plota o gráfico com os dados gerados
    plt.bar(x, dados)
    
    # Aguarda um tempo determinado
    plt.pause(0.2)
# Delisga o método interativo 
plt.ioff()    
plt.show
```
