```python
import numpy as np
x = np.arange(0,10, 0.1)
figl, fl_axes = plt.subplots(ncols=2, nrows=2, figsize=(15,10))
figl.suptitle('Triconometria', size=30)

fl_axes[0, 0].plot(np.sin(x))
fl_axes[0, 0].set_title('Seno')
fl_axes[0, 0].set_xlabel('Seno - Eixo x')
fl_axes[0, 0].set_ylabel('Seno - Eixo y')


fl_axes[0, 1].plot(np.cos(x))
fl_axes[0, 1].set_title('Cosseno')
fl_axes[0, 1].set_xlabel('Cosseno - Eixo x')
fl_axes[0, 1].set_ylabel('Cosseno - Eixo y')


fl_axes[1, 0].plot(np.tan(x))
fl_axes[1, 0].set_title('Tangente')
fl_axes[1, 0].set_xlabel('Tangente - Eixo x')
fl_axes[1, 0].set_ylabel('Tangente - Eixo y')

fl_axes[1, 1].plot(np.sin(x), label='Seno', color='purple')
fl_axes[1, 1].plot(np.cos(x), label='Cosseno', color='red')
fl_axes[1, 1].set_title('Seno e Cosseno')
fl_axes[1, 1].set_xlabel('Seno e Cosseno - Eixo x')
fl_axes[1, 1].set_ylabel('Seno e Cosseno - Eixo y')
fl_axes[1, 1].legend()
```
