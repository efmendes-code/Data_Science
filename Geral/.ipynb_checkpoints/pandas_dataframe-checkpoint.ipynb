# Pandas DataFrame

Pandas DataFrame  uma estrutura de dados bidimensional com os dados alinhados de forma tabular em linhas e colunas, mutvel em tamanho e potencialmente heterognea, semelhantemente a uma pasta de trabalho do MS-EXCEL. 

A diferena essencial  que os nomes de colunas e os nmeros de linha so conhecidos como ndice de coluna e linha, no caso do DataFrame. 

As colunas possuem nomes (ndice da coluna) e, as linhas, podem ter nomes referentes a colunas e as linhas podem ter nomes (ndices textuais) ou podem, por padro, ser numeradas (ndice numrico).





# Criando um DataFrame

Em geral, o DataFrame pode conter dados a partir de:

Um DataFrame do Pandas

```python
#Importando a biblioteca Pandas
import pandas as pd
```

```python
#Criando um Dataframe com 2 linhas (indexes 0 e 1) e 4 colunas ('Idade', 'Sexo', 'Peso', 'Altura').
meu_df = pd.DataFrame([[21,'F', 50, 1.57],[22,'F',58, 1.70]], index=range(0,2), columns=['Idade', 'Sexo', 'Peso', 'Altura'])
```

```python
#Exibindo o DataFrame criado
meu_df
```

```python
#Acesso as linhas do DataFrame a partir do seu ndice. Utilizar a funo .iloc[indice].
#Exibir a segunda linha do Dataframe
meu_df.iloc[1]
```

Uma Srie Pandas: um array unidimensional capaz de armazenar qualquer tipo de dados com rtulos ou ndice de eixo. 

Um exemplo de um objeto Series  uma coluna de um DataFrame.

```python
#Criando uma Srie formada por nomes e com ndices de 0 at 4.
nomes = pd.Series(["Luciano","Matheus", "Rodrigo", "Bruno", "Michel"], index=range(0,5))
nomes
```

```python
#Criando um DataFrame a partir da Serie nomes com a coluna denominada Nomes    
df = pd.DataFrame(data=nomes, columns=['Nomes'])
```
